```python
#importing pandas to read csv 
import pandas as pd

duration = pd.read_csv('netflix_data.csv')

movies = duration[duration['type'] == 'Movie'].copy()

#most frequent movie duration in the 1990s
movies_90s = movies[(movies['release_year'] >= 1990) & (movies['release_year'] < 2000)]

#setting duration to integer
movies_90s['duration'] = movies_90s['duration'].astype(int)
duration = int(movies_90s['duration'].mode()[0])

action_movies = movies_90s[movies_90s['genre'] == 'Action']

#short movie duration (90 mins) count
short_movie_count = (action_movies['duration'] < 90).sum()
print('most_frequent_movies_90s_duration: ', duration)
print('short_action_movies: ', short_movie_count)

#matplotlip for visualization
import matplotlib.pyplot as plt

plt.hist(movies_90s["duration"])
plt.title('Distribution of Movie Durations in the 1990s')
plt.xlabel('Duration (minutes)')
plt.ylabel('Number of Movies')
plt.show()

duration = 100
```
